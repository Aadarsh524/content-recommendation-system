```python
import pandas as pd
```

```python
#Ratings data
rating = pd.read_csv('../data/processed/rating.csv')
rating.head()
```

```python
rating = rating.sort_values("timestamp")
rating.head()
```

```python
from sklearn.model_selection import train_test_split

train_rows = []
valid_rows = []
test_rows = []

for user, group in rating.groupby("user_id"):
    group = group.sort_values("timestamp")

    n = len(group)
    test_size = int(0.2 * n)
    val_size = int(0.2 * n)

    test = group.iloc[-test_size:]
    valid = group.iloc[-(test_size + val_size):-test_size]
    train = group.iloc[:-(test_size + val_size)]

    train_rows.append(train)
    valid_rows.append(valid)
    test_rows.append(test)

train_df = pd.concat(train_rows)
valid_df = pd.concat(valid_rows)
test_df = pd.concat(test_rows)
```

```python

import pickle

with open("../models/train_valid_test.pkl", "wb") as f:
    pickle.dump((train_df, valid_df, test_df), f)
```
