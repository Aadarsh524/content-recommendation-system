```python
#Types of genres
genre = pd.read_csv('../data/raw/u.genre',sep="|",encoding='latin-1', header=None)
genre.columns=['Genre', 'Number']
genre.drop(index=0, axis=1, inplace=True)
genre_list = list(genre['Genre'])
genre_list
```

```python
#Types of occupations
occupation = pd.read_csv('../data/raw/u.occupation', sep="|", encoding='latin-1', header=None)
occupation.columns = ['Occupations']
occupation_list = list(occupation['Occupations'])
occupation_list
```
