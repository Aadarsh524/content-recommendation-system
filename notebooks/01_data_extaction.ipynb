```python
import pandas as pd
```

```python
#Information about the data
info = pd.read_csv('../data/raw/u.info', sep=" ", header=None)
info.columns=['Counts', 'Type']
info
info.to_csv('../data/processed/info.csv', index=False)
```

```python
#Types of genres
genres = pd.read_csv('../data/raw/u.genre', sep="|", encoding='latin-1', header=None)
genres.columns = ['genre', 'number']
genres = genres.dropna()              # drop last row which is empty
genres = genres[genres['genre'] != ''] 

genre_list = list(genres['genre'])
genre_list
genres.to_csv('../data/processed/genre.csv', index=False)
```

```python
#Types of occupations
occupations = pd.read_csv('../data/raw/u.occupation', sep="|", encoding='latin-1', header=None)
occupations.columns = ['occupation']
occupation_list = list(occupations['occupation'])
occupation_list
occupations.to_csv('../data/processed/occupation.csv', index=False)
```

```python
#Ratings data
rating = pd.read_csv('../data/raw/u.data', sep="\t", header=None)
rating.columns = ['user_id', 'item_id', 'rating', 'timestamp']
rating.head()
rating.to_csv('../data/processed/rating.csv', index=False)
```

```python
#user info
users = pd.read_csv('../data/raw/u.user', sep="|", header=None)
users.columns=['user_id','age', 'gender', 'occupation', 'zip']
users
users.to_csv('../data/processed/user.csv', index=False)
```

```python
#Movies data
items = pd.read_csv('../data/raw/u.item', sep="|", encoding='latin-1', header=None)
items.columns = ['item_id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 
                'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 
                'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items.head()
items.to_csv('../data/processed/item.csv', index=False)
```
